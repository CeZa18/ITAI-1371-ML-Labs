## Part 2
Loading and exploring the first Dataset by using a classic Dataset that contains measurements of iris flowers from three different species.

```python
# Load a simple dataset (Iris flowers - classic beginner dataset)
from sklearn.datasets import load_iris

# Load the data
iris = load_iris()
print("Dataset loaded successfully!")
print(f"Dataset shape: {iris.data.shape}")
print(f"Features: {iris.feature_names}")
print(f"Target classes: {iris.target_names}")
```

```python
# Convert to pandas DataFrame for easier handling
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

# Display the first few rows
print("First 5 rows of the dataset:")
print(df.head())

print("\nDataset info:")
print(df.info())
```

## Part 3
Create a data visualization crucial for machine learning.

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))

# Plot sepal length vs sepal width, colored by species
species_colors = {'setosa': 'red', 'versicolor': 'blue', 'virginica': 'green'}

for species in df['species'].unique():
  species_data = df[df['species'] == species]
  plt.scatter(species_data['sepal length (cm)'],
              species_data['sepal width (cm)'],
              c=species_colors[species],
              label=species,
              alpha=0.7)

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset: Sepal Length vs Sepal Width')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(" 👏 Congratulations! You've created your first data visualization!")
```

```python
# Basic statistical analysis
print("Basic Statistics for Iris Dataset:")
print("=" * 40)

# Calculate mean values for each species
species_means = df.groupby('species').mean()
print("\nMean values by Species:")
print(species_means)

# Added an additional calculation to show the standard deviation for each species
species_std = df.groupby('species').std()
print("\nStandard Deviation by Species:")
print(species_std)

# Count samples per species
species_counts = df['species'].value_counts()
print("\nSamples per Species:")
print(species_counts)
```

```python
# Task 1: Create a simple calculation using NumPy
# Calculate the mean and the standard deviation of sepal length

sepal_length = df['sepal length (cm)']

# Are we suppose to write our own version of code to calculate mean and std?
mean_sepal_length = np.mean(sepal_length) # Calculate the mean of sepal lengths
std_sepal_length = np.std(sepal_length) # Calculate the standard deviation of sepal lengths

print(f"Mean of Sepal length: {mean_sepal_length:.2f} cm")
print(f"Standard deviation: {std_sepal_length:.2f} cm")

assert isinstance(mean_sepal_length, (float, np.floating)), "Mean should be a number"
assert isinstance(std_sepal_length, (float, np.floating)), "Standard deviation should be a number"

print("✔️ Task 1 completed successfully!")
```

```python
# Task 2: Create a bar chart showing species counts
species_counts = df['species'].value_counts()

plt.figure(figsize=(8, 5))
plt.bar(species_counts.index, species_counts.values, color=['red', 'blue', 'green'])
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Species Counts in Iris Dataset')
plt.show()

print(f"Species distribution: {dict(species_counts)}")
print("✔️ Task 2 completed successfully!")
```

```python
# Practice more visualizations by adding best fit line using a trending line plot

# 1) Set style and create the plot
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

# Format species colors
species_colors = {'setosa': 'red', 'versicolor': 'blue', 'virginica': 'green'}

# 2) Plot the trendlines for each species 2.1) Use sns.regplot to automatically
# plot data points and the best fit trendline
for species in df['species'].unique():
  species_data = df[df['species'] == species]
  sns.regplot(x='sepal length (cm)', y='sepal width (cm)', data=species_data,
              color=species_colors[species])

# 3) Labels and title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Best Fit Line: Sepal Length vs Sepal Width')

# 4) Show the plot
plt.show()
```
